```python
# Initial imports
import pandas as pd
from pathlib import Path
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

## Data Preparation

```python
# Data loading
file_path = Path("Resources/myopia.csv")
df_myopia = pd.read_csv(file_path)
df_myopia.head(10)
```

```python
# List dataframe data types
df_myopia.dtypes
```

```python
# Removing the MYOPIC  Column
df_myopia = df_myopia.drop(columns=["MYOPIC"])
df_myopia
```

```python
# Find null values
for column in df_myopia.columns:
    print(f"Column {column} has {df_myopia[column].isnull().sum()} null values")
```

```python
# Find duplicate entries
print(f"Duplicate entries: {df_myopia.duplicated().sum()}")
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_myopia[['AGE', 'SPHEQ', 'AL', 'ACD', 'LT', 'VCD', 'SPORTHR', 'READHR', 'COMPHR', 'STUDYHR', 'TVHR', 'DIOPTERHR']])
```

```python
df_myopia.columns
```

```python
# Create a DataFrame with the transformed data
new_df_myopia = pd.DataFrame(scaled_data, columns=df_myopia.columns[2:])
new_df_myopia['MOMMY'] = df_myopia['MOMMY']
new_df_myopia['DADMY'] = df_myopia['DADMY']
new_df_myopia.head()
```

## PCA Model

```python
# Initialise PCA model for 3 principal components
pca = PCA(n_components=0.90)

# Get two principal components for the iris data.
myopia_pca = pca.fit_transform(new_df_myopia)
```

```python
# Transform PCA data to a DataFrame
df_myopia_pca = pd.DataFrame(
    data=myopia_pca, columns=["principal component 1", "principal component 2","principal component 3","principal component 4","principal component 5","principal component 6","principal component 7","principal component 8"]
)
df_myopia_pca.head()
```

## t-SNE Model

```python
# Run t-SNE
tsne = TSNE(learning_rate=250)
tsne_features = tsne.fit_transform(df_myopia_pca)
```

```python
# Check that dimensions have been reduced to two
tsne_features.shape
```

```python
# Plot the results
plt.scatter(tsne_features[:,0], tsne_features[:,1])
plt.show()
```

## K-Means Model

```python
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(df_myopia_pca)
    inertia.append(km.inertia_)

# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow
```

```python
# Plot the elbow curve to find the best candidate(s) for k
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow curve for customer data')
plt.show()
```

## Conclusion

From the above Analysis, we can safely conclude that we can cluster the data into two groups, based on the tsne plot where two distinct clusters are visible.
